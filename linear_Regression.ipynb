```python

df['Total_spent'] = df['TV'] + df['Radio'] + df['Newspaper']
df.head()
```

```python
x= df.drop(['Sales','Total_spent'],axis=1)
x
```

```python
y=df['Sales']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=101)
x_train.head()
```

```python
len(x_train)
```

```python
len(df)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
test_pred=model.predict(x_test)
test_pred
```

```python

from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
sns.histplot(data=df,x='Sales',bins=20)
```

```python
mean_absolute_error(y_test,test_pred)
```

```python
z=mean_squared_error(y_test,test_pred)
```

```python
np.sqrt(z)
"""If there is lot of difference between MAE and RMSE then it means your model is performing well on most of the points nearby but not far away points.
"""
```

```python
test_residuals = y_test - test_pred
test_residuals
```

```python
sns.scatterplot(x=y_test, y=test_residuals)
#it looks like random, no pattern in the residual plot can be seen so linear regression can be used
plt.axhline(y=0, color='red', ls='--')
```

```python
sns.displot(test_residuals, bins=25, kde=True) #kernel density estimation
```

```python
model.coef_   #theta coefficients for different features, newspaper coefficient is almost zero so it dosen't contribute much for the sales(it will reduce the sales),
#radio coeffiecient is the strongest, if you spend on it then it will show increase in sales 0.185 times, keeping TV and newspaper expenditure constant
```

```python
y_hat = model.predict(x)


```

```python
fig,axes = plt.subplots(nrows=1, ncols=3, figsize=(16,6))

axes[0].plot(df['TV'], df['Sales'], 'o')
axes[0].plot(df['TV'], y_hat, 'o', color='red')
axes[0].set_ylabel("Sales")
axes[0].set_title("TV Spend")

axes[1].plot(df['Radio'], df['Sales'], 'o')
axes[1].plot(df['Radio'],y_hat, 'o', color='red')
axes[1].set_ylabel("Sales")
axes[1].set_title("Radio Spend")

axes[2].plot(df['Newspaper'], df['Sales'], 'o')
axes[2].plot(df['Newspaper'],y_hat, 'o', color='red')
axes[2].set_ylabel("Sales")
axes[2].set_title("Newspaper spend")
```
